Abaixo será testado o modelo de decision trees

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
variaveis = [   'Elevation','Aspect', 'Slope', 'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Hydrology', 'Horizontal_Distance_To_Fire_Points',
                "Horizontal_Distance_To_Roadways",'Hillshade_3pm','Wilderness_Area2','Wilderness_Area3','Wilderness_Area4',"Soil_Type1", "Soil_Type2", "Soil_Type5",
                "Soil_Type6", "Soil_Type10","Soil_Type11",  "Soil_Type13", "Soil_Type14", "Soil_Type16", "Soil_Type17", "Soil_Type18", "Soil_Type30",
                "Soil_Type35", "Soil_Type36", "Soil_Type37", "Soil_Type38", "Soil_Type39", "Soil_Type40" ]
```

```python
x = dt[variaveis]
y = dt["Cover_Type"]
```

```python
x_prev = dt[variaveis]
```

Abaixo será feito arquivos falsos baseados nos dados originais, em cada seed será feito novos arquivos aleatórios para teste. Desta forma é possível testar a a média de acertos.

```python
dados_falsos = np.arange(50000)
dados_falsos
```

```python
np.random.seed()   #Ter sempre o mesmo valor no teste, memso sendo aleatório

x_treino, x_valid, y_treino, y_valid = train_test_split(x, y, test_size=0.5)
```

```python
modelo = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=0)
modelo.fit(x_treino, y_treino)
```

```python
a = modelo.predict(x_valid)
np.mean(y_valid == a)
```

Após criar a arvore de decisão, será testado no arquivo "test"


```python
x_prev = test[variaveis]
p = modelo.predict(x_prev)
```

```python
prev_final = pd.Series(p, index=test['Id'], name="Cover_Type")
```
